```python
from scipy.stats import cauchy
```

```python
from scipy.optimize import minimize
```

```python
theta_points = np.arange(-0.04, -0.01, 10000)
```

```python
def max_arg(theta):
    x = cauchy.rvs(loc=theta, scale=abs(theta), size=10000)
    return np.sum(np.log(cauchy.pdf(x, loc=theta, scale=abs(theta))))
```

```python
max_arg(-0.01)
```

```python
theta_results = np.array([max_arg(theta) for theta in theta_points])
```

```python
df = pd.DataFrame(data={'theta': theta_points, 'result': theta_results})
```

```python
df[df.result == df.result.max()]
```

```python
from scipy.stats import lognorm
```

```python
def max_arg(theta):
    x = lognorm.rvs(s=abs(theta-0.01), loc=theta, scale=np.exp(theta), size=10000)
    return np.sum(np.log(lognorm.pdf(x=x, s=abs(theta-0.01), loc=theta, scale=np.exp(theta))))
```

```python
theta_results = np.array([max_arg(theta) for theta in [0.02, 0.03]])
```

```python
theta_results
```
