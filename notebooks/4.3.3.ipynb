```python
import numpy as np
from numpy.random import multivariate_normal
from statsmodels.sandbox.tools.tools_pca import pca
```

```python
N = 50
```

```python
# Generate location vector
mu = np.zeros((N,))
mu
```

```python
# Generate positive definite scatter matrix
sigma = np.identity(N)
sigma
```

```python
T = 20

# Create list of samples
sample = multivariate_normal(mu, sigma, T)

# Calculate covariance matrix
sample_covariance = np.cov(sample)

# Perform PCA on sample covariance matrix
evals = pca(sample_covariance)[2]

evals
```

```python
num_experiments = 100
```

```python
evals_dict = {}
evals_avg_dict = {}

for T in xrange(2, 20):
    evals_dict[T] = np.zeros((N, num_experiments))
    for i in xrange(num_experiments):
        # Create list of samples
        sample = multivariate_normal(mu, sigma, T)
        
        # Calculate covariance matrix
        sample_covariance = np.cov(sample.T)
        
        # Perform PCA on sample covariance matrix
        evals = pca(sample_covariance)[2]
        
        # Store eigenvalues for each i
        evals_dict[T][:,i] = evals
    
    evals_avg_dict[T] = np.average(evals_dict[T], axis=1)
```

```python
X = np.arange(1, 51)
Y = np.arange(2, 20)
X, Y = np.meshgrid(X, Y)
```

```python
Z = np.zeros((18, 50))
for y in Y:
    Z[y[0]-2,:] = evals_avg_dict[y[0]]
```
